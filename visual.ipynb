```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart=pd.read_excel(r"C:\Users\user\Desktop\New folder\mart_train_sample.xlsx") 
mart
```

```python
sns.countplot(data=mart, x='Outlet_Size')
```

```python
sns.barplot(data=mart, x='Outlet_Size', y='Sales', estimator='mean')
```

```python
sns.boxplot(data=mart, x='Outlet_Size', y='Sales')
plt.title('Sales Distribution by Outlet Size')
```

```python
sns.violinplot(data=mart, x='Outlet_Size', y='Sales')
plt.title('Sales Distribution and Density by Outlet Size')
```

```python
sns.stripplot(data=mart, x='Outlet_Size', y='Sales', jitter=True)
plt.title('Sales Data Points by Outlet Size')
```

```python
sns.swarmplot(data=mart, x='Outlet_Size', y='Sales')
plt.title('Sales Data Points by Outlet Size (Swarm Plot)')
```

```python
sns.catplot(data=mart, x='Outlet_Size', y='Sales', kind='box')
```

```python
sns.histplot(data=mart, x='Sales', bins=20)
```

```python
sns.kdeplot(data=mart, x='Sales')
```

```python
sns.rugplot(data=mart, x='Sales', color='r')
```

```python
sns.ecdfplot(data=mart, x='Sales')
```

```python
sns.displot(data=mart, x='Sales', kde=True, rug=True)
```

```python
sns.jointplot(data=mart, x='Item_MRP', y='Sales', kind='scatter')
```

```python
sns.pairplot(mart)
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales')
```

```python
sns.relplot(data=mart, x='Outlet_Year', y='Sales', kind='line')
```
